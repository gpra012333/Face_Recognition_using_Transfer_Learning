```python
import numpy as np
def get_label(image_name):
    name = image_name.split('.')[-3]
    if(name=='Gaurav'):
        return np.array([1,0,0])
    if(name=='Paati'):
        return np.array([0,1,0])
    if(name=='Uday'):
        return np.array([0,0,1])
```

```python
import os
from random import shuffle
from tqdm import tqdm
```

```python
import cv2
def my_data():
    data = []
    for img in tqdm(os.listdir("data")):
        path = os.path.join("data", img)
        img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img_data = cv2.resize(img_data, (50,50))
        data.append([np.array(img_data), get_label(img)])
    shuffle(data)
    return data
```

```python
data = my_data()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
train_x = train_X/255.0
test_x = test_X/255.0
```

```python
import tensorflow as tf
import tensorflow.keras.layers as tfl
from tensorflow import keras
```

```python
cnn_model= keras.models.Sequential([
    tfl.Conv2D(filters=64, kernel_size=3, activation='relu', input_shape= (50,50,1)),
    tfl.MaxPooling2D(pool_size=2),
    
    tfl.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape= (50,50,1)),
    tfl.MaxPooling2D(pool_size=2),
    
    tfl.Flatten(),
    tfl.Dense(2024, activation='relu'),
    tfl.Dropout(0.5),
    
    tfl.Dense(1024, activation='relu'),
    tfl.Dropout(0.5),
    
    tfl.Dense(512, activation='relu'),
    tfl.Dropout(0.5),
    tfl.Dense(3, activation='softmax')  
])

```

```python
cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
cnn_model.fit(train_x, train_Y, epochs=1)
```

```python
cnn_model.evaluate(test_x, test_Y)
```

```python
import cv2
import numpy as np

def get_predict(model):
    # Load the face classifier
    face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    
    def croped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        if len(faces) == 0:
            return None, None
        # Assuming there's only one face in the frame, you might want to handle multiple faces differently
        (x, y, w, h) = faces[0]
        crop_img = gray[y:y+h, x:x+w]
        return crop_img, (x, y, w, h)

    cap = cv2.VideoCapture(0)

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        face, bbox = croped(frame)
        if face is not None:
            face = cv2.resize(face, (50, 50))
            face = np.expand_dims(face, axis=-1)  # Add channel dimension
            face = np.expand_dims(face, axis=0)   # Add batch dimension
            face = face / 255.0
            prediction = model.predict(face)
            label = ["Gaurav", "Paati", "Uday"]
            predicted_label = np.argmax(prediction, axis=1)[0]  # Get the index of the highest probability
            print(f"Predicted class label: {label[predicted_label]}")

            # Draw rectangle around detected face
            (x, y, w, h) = bbox
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            # Put prediction text on the frame
            cv2.putText(frame, f"Prediction: {label[predicted_label]}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

        cv2.imshow('Face', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Assuming you have already trained your CNN model and stored it in 'cnn_model'
# Call the function with your model
get_predict(cnn_model)
```
